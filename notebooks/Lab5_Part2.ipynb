```python
import pyspark  
from pyspark.mllib.regression import LabeledPoint
from pyspark.mllib.classification import LogisticRegressionWithSGD
from pyspark.mllib.tree import DecisionTree
import re
```

```python
import itertools
import csv
```

```python
sc = pyspark.SparkContext.getOrCreate()
```

```python
lemma = sc.textFile("new_lemmatizer.csv")\
        .map(lambda x: (x.split(",")[0],list(filter(None,x.split(",")[1::]))))\
```

```python
lemma_out = lemma.collect()
```

```python
lemma_out
```

```python
lemma_dict=dict(lemma_out)
```

N-gram = 2

```python
text1 = sc.textFile("lucan.bellum_civile.part.1.tess")\
        .map(lambda x: ((x.split("\t")[0].replace("<", "")).replace(">", ""),list(filter(None,re.findall(r"[\w']+",x.split("\t")[1].lower().replace("j","i").replace("v","u"))))))\
        .map(lambda x: (x[0],list(itertools.combinations(x[1],2))))\
        .flatMap(lambda x: [(values,x[0]) for values in x[1]])
```

```python
text1_out = text1.collect()
```

```python
text1_out
```

```python
text2 = sc.textFile("vergil.aeneid.tess")\
        .filter(lambda x: x is not "")\
        .map(lambda x: ((x.split("\t")[0].replace("<", "")).replace(">", ""),list(filter(None,re.findall(r"[\w']+",x.split("\t")[1].lower().replace("j","i").replace("v","u"))))))\
        .map(lambda x: (x[0],list(itertools.combinations(x[1],2))))\
        .flatMap(lambda x: [(values,x[0]) for values in x[1]])
```

```python
text2_out = text2.collect()
text2_out
```

```python
textmerge = text1_out + text2_out
```

```python
textmerge_out = []
for x in textmerge:
    if lemma_dict.get(x[0][0]):
        for pair1 in lemma_dict.get(x[0][0]):
            if lemma_dict.get(x[0][1]):
                for pair2 in lemma_dict.get(x[0][1]):
                    textmerge_out.append(((pair1,pair2),x[1]))
            else:
                textmerge_out.append(((pair1,x[0][1]),x[1]))
    else:
        if lemma_dict.get(x[0][1]):
            for pair2 in lemma_dict.get(x[0][1]):
                textmerge_out.append(((x[0][0],pair2),x[1]))
        else:
            textmerge_out.append(((x[0][0],x[0][1]),x[1]))
```

```python
textmerge_out
```

```python
from collections import defaultdict

data_dict = defaultdict(list)
```

```python
for x in textmerge_out:
    data_dict[x[0]].append(x[1])
```

for x in data_dict:
    print(x,end=" "),
    print(": ",end=" "),
    print(data_dict.get(x))

```python
with open('ngram_2.csv', "w", newline="\n", encoding="utf-8") as csv_file:
    writer = csv.writer(csv_file)
    for key, value in data_dict.items():
       writer.writerow([key, value])
```

```python
text4 = sc.textFile("vergil.aeneid.tess")\
        .filter(lambda x: x is not "")\
        .map(lambda x: ((x.split("\t")[0].replace("<", "")).replace(">", ""),list(filter(None,re.findall(r"[\w']+",x.split("\t")[1].lower().replace("j","i").replace("v","u"))))))\
        .map(lambda x: (x[0],list(itertools.combinations(x[1],3))))\
        .flatMap(lambda x: [(values,x[0]) for values in x[1]])
```

```python
text4_out = text4.collect()
text4_out
```

```python
textmerge_2 = text3_out + text4_out
```

```python
textmerge_2_out = []
for x in textmerge_2:
    if lemma_dict.get(x[0][0]):
        for pair1 in lemma_dict.get(x[0][0]):
            if lemma_dict.get(x[0][1]):
                for pair2 in lemma_dict.get(x[0][1]):
                    if lemma_dict.get(x[0][2]):
                        for pair3 in lemma_dict.get(x[0][2]):
                            textmerge_2_out.append(((pair1,pair2,pair3),x[1]))
                    else:
                        textmerge_2_out.append(((pair1,pair2,x[0][2]),x[1]))
                    
            else:
                if lemma_dict.get(x[0][2]):
                    for pair3 in lemma_dict.get(x[0][2]):
                        textmerge_2_out.append(((pair1,x[0][1],pair3),x[1]))
                else:
                    textmerge_2_out.append(((pair1,x[0][1],x[0][2]),x[1]))
    else:
        if lemma_dict.get(x[0][1]):
            for pair2 in lemma_dict.get(x[0][1]):
                if lemma_dict.get(x[0][2]):
                    for pair3 in lemma_dict.get(x[0][2]):
                        textmerge_2_out.append(((x[0][0],pair2,pair3),x[1]))
                else:
                    textmerge_2_out.append(((x[0][0],pair2,x[0][2]),x[1]))
                    
        else:
            if lemma_dict.get(x[0][2]):
                for pair3 in lemma_dict.get(x[0][2]):
                    textmerge_2_out.append(((x[0][0],x[0][1],pair3),x[1]))
            else:
                textmerge_2_out.append(((x[0][0],x[0][1],x[0][2]),x[1]))
```
